# DATA CLEANING

```python
carprice_data.isnull() #Count the number of missing values in data frame 
```

```python
carprice_data.isnull().sum() #Count the number of missing values in each column
```

```python
threshold = len(carprice_data)* 0.1 #Remove columns that are 90% empty OR 10% of my rows )
threshold
```

```python
carprice_data.dropna(thresh=threshold, axis = 1, inplace = True)
```

```python
print(carprice_data.isnull().sum())
```

# DATA IMPUTATION AND MANIPULATION

```python
def impute_median(series): #Define a function impute_median
    return series.fillna(series.median())
```

# DATA VISUALIZATION

```python
grp = carprice_data.groupby('CarName')
x = grp['wheelbase'].agg(np.mean)
y = grp['price'].agg(np.sum)
z = grp['carlength'].agg(np.mean)
print(x)
print(y)
print(z)
```

```python
plt.figure(figsize=(17,4))
plt.plot(x, 'ro', color='r')
plt.xticks(rotation=90)
plt.xlabel('Categories-->')
plt.ylabel('Rating-->')
plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(y, 'r--', color='b')
plt.xticks(rotation=90)
plt.title('Category wise Pricing')
plt.xlabel('Categories-->')
plt.ylabel('Rating-->')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(z, 'g^', color='g')
plt.xticks(rotation=90)
plt.title('Category wise Pricing')
plt.xlabel('Categories-->')
plt.ylabel('Rating-->')
plt.show()
```
